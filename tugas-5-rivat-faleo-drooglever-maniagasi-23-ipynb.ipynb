```python
data = [-10,-30,-50,-80,100,80,50,30,10]
terbesar = data [0]
data_terurut = sorted(data)
print(data_terurut)
```

```python
for i in range(len(data)):
    if data[i] > terbesar:
        terbesar = data[i]
print(terbesar)
```

```python
nilai_positif = []
for angka in data:
    if angka > 0:
        nilai_positif.append(angka)
print("Nilai_positif:", nilai_positif)

nilai_negatif = []
for angka in data:
    if angka < 0:
        nilai_negatif.append(angka)
print("Nilai_negatif:", nilai_negatif)
```
